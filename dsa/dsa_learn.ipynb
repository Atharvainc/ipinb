```python
def selec_sort(nums:list)->list:
    n=len(nums)
    for i in range(n):
        min_idx=i
        for j in range(i+1,n,1):
            if nums[j]<nums[min_idx]:
                min_idx=j
        nums[i],nums[min_idx]=nums[min_idx],nums[i]
    return nums
selec_sort(nums)
#TC=O(n^2)
#SC=O(1)
```

```python
nums=[5,7,8,4,1,6,9,2]
```

```python
def bub_sort(nums:list)->list:
    for i in range(len(nums)-2,-1,-1):
        for j in range(0,i+1):
            if nums[j]>nums[j+1]:
                nums[j],nums[j+1]=nums[j+1],nums[j]
    return nums
bub_sort(nums)
#TC=O(n^2)
#SC=O(1)
```

#### quik sort

```python
def partition(nums, low, high):
    pivot = nums[low]
    i = low
    j = high
    
    while i < j:
        while nums[i] <= pivot and i < high: 
            i += 1
        while nums[j] > pivot and j > low: 
            j -= 1
        if i < j:
            nums[i], nums[j] = nums[j], nums[i]
    nums[low], nums[j] = nums[j], nums[low]
    return j

def quik_sort(nums, low, high):
    if low < high:
        a = partition(nums, low, high)
        quik_sort(nums, low, a - 1)
        quik_sort(nums, a + 1, high)

nums = [3, 1, 2, 4, 1, 5, 2, 6, 4]
quik_sort(nums, 0, len(nums) - 1)
print(nums)
```

```python
#longest consecutive sequence
nums=[1,99,101,98,2,5,3,100,1,1]
def lcs(nums:list)->int:
    nset=set(nums)
    gmax=0
    for n in nset:
        if n-1 not in nset:
            x=n
            c=1
            while x+1 in nset:
                c+=1
                x+=1
            gmax=max(gmax,c)
    return gmax
# we use set for O(1) search
lcs(nums)
```

## doubly linked list

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        """Add an element to the end of the list."""
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    def display_forward(self):
        """Print the list from start to finish."""
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" <-> ".join(elements))

    def display_backward(self):
        """Print the list from finish to start."""
        current = self.head
        if not current: return
        while current.next:
            current = current.next
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.prev
        print(" <-> ".join(elements))
```

```python
#generating all subsequences with sum=k
nums = [5, 9, 4]
res = []

def subsumequalk(idx, subs, current_sum, target):
    # Base Case: If we've considered all elements
    if idx >= len(nums):
        if current_sum == target:
            res.append(subs.copy())
        return
#    elif current_sum>target:#only if there are all +ve nos in list
#       return
    # Choice 1: PICK the current element
    subs.append(nums[idx])
    subsumequalk(idx + 1, subs, current_sum + nums[idx], target)
    
    # Choice 2: DON'T PICK the current element (Backtrack)
    subs.pop()
    subsumequalk(idx + 1, subs, current_sum, target)

# Start from index 0
subsumequalk(0, [], 0, 9)
print(res) # Output: [[5, 4], [9]]
```

```python
nums = [5, 9, 4]

def exists_subsequence(idx, current_sum, target):
    # 1. Base Case: Success!
    if current_sum == target:
        return True
    
    # 2. Base Case: Failure (End of list or exceeded sum)
    if idx >= len(nums):
        return False

    # 3. Backtracking Step 1: PICK the current number
    # If this path returns True, we stop and pass True all the way up
    if exists_subsequence(idx + 1, current_sum + nums[idx], target):
        return True
    
    # 4. Backtracking Step 2: DON'T PICK the current number
    # If the first path failed, try the second path
    if exists_subsequence(idx + 1, current_sum, target):
        return True

    # 5. If neither path worked, this branch is False
    return False

print(exists_subsequence(0, 0, 9))
```
