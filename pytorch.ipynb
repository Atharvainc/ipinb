#### we will use a linear regression to make datasets with known parameters 

```python
wt=0.7
bias=0.3

start,end,step=0,1,0.02
x=tc.arange(start,end,step).unsqueeze(dim=1)
y=wt*x+bias
x[:10],y[:10]
```

### creating training and testing datasets

```python
len(x)
```

```python
split=int(0.8*len(x))
xtrain,xtest,ytrain,ytest=x[:split],x[split:],y[:split],y[split:]
```

```python
plt.plot(x,y)
plt.show()
```

## building a model
###### it is a lrm 
__y=ax+b__  

##### how does a model work?
- start with random values
- look at training data,create a model and fit the test data accordingly
- how does it do so??
    - thru algo's like gradient descent and back propogation

##### what our model does?
- Start with random wt and bias values.
- look at training data and adjust the random values to better represent(fit) the ideal values.

##### how does it do?
- two main algo's:
    - gradient descent(uses requires_grad param)
    - backpropogation

### model building essentials
- torch.nn: contains all the building blocks for computational graphs(a neural net can be considered a computational graph).
- torch.nn.Parameter: what parameters should our model try and learn, often a pytorch layer from torch.nn will set these for us.
- torch.nn.Module: the base class for all neural network modules, if you subclass it, you should overwrite forward()
- torch.optim: this is where the optimizers in pytorch live, they will help with gradient descent.
- def forward(): all nn. Module subclasses require you to overwrite forward(), this method defines hat happens in the forward computation.


```python
class lrm(nn.Module):#nn.Module has all the building blocks for nn
    def __init__(self):
        super().__init__()
        # start with random weight and bias and try to reach the ideal weight and bias
        self.wt=nn.Parameter(tc.randn(1,requires_grad=True,dtype=tc.float32))
        self.b=nn.Parameter(tc.randn(1,requires_grad=True,dtype=tc.float32))
    #forward method to define the computation in the model
    def forward(self,x:tc.Tensor)->tc.Tensor:
        return self.wt*x+self.b
```

### we can check model parameters using .parameters()

```python
seed=42
tc.manual_seed(seed)
model0=lrm()
list(model0.parameters())
```

```python
[wt,bias]
```

### making predictions using tc.inferenc_mode()

- disables all the benifits, helping save data.

```python
with tc.inference_mode():
    ypred=model0(xtest)
ypred,ytest
```

### train model

##### loss function(used to measure the wrongness of model),(lower the better)
##### optimizer(takes into account the loss of a model and adjusts the model's parameters to improve loss fn.


```python
# loss fn setup
loss_fn=nn.L1Loss()#mean absolute error
# optimizer setup
optimizer=tc.optim.SGD(params=model0.parameters(),#optimizing parameters present in the model
                       lr=0.01)#learning rate(lesser=slower, higher=risk of losing data(knowledge))
```

### building a training loop

#### things needed in a training loop
0. loop thru data.
1. forward pass/propogation(this involves data moving thru our model's `forward` fn.) to make predictions on data.
2. calculate the loss fn(real vs predicted).
3. optimizer zero grad.
4. loss backward: move backwards thru the nn to calculate the gradients of each of the parameters of our model with respect to the loss.**(back propogation)**
5. optimizer step:  use the optimizer to adjust our model's parameters to try and improve the loss.**(gradient descent)**

```python
'''
epochs=1#epoch is one loop thru the data(hyperparameter)

#0. 
for epoch in range(epochs):
    # set model to training mode
    model0.train()# train mode in torch sets all params that require grad to require grad
    
    #1.
    ypred=model0(xtrain)
    #2.
    loss=loss_fn(ypred,ytrain)
    #3.
    optimizer.zero_grad()
    #4.
    loss.backward()
    #5.
    # by default how the optimizer changes will accumulate thru the loop so we have to zero them above in 3. for the next iter of the loop.
    optimizer.step()

    model0.eval()#eval mode disables grad tracking
'''    
```

### testing loop

```python
tc.manual_seed(seed)
epochs = 200
epoch_count, loss_vals, test_loss_vals = [], [], []

for epoch in range(epochs):
    model0.train()
    optimizer.zero_grad()     # 1. Zero gradients first ✅
    ypred = model0(xtrain)    # 2. Forward pass
    loss = loss_fn(ypred, ytrain)  # 3. Compute loss
    loss.backward()           # 4. Backward pass
    optimizer.step()          # 5. Update weights

    if epoch % 10 == 0:
        model0.eval()
        with tc.inference_mode():
            ycap = model0(xtest)
            test_loss = loss_fn(ycap, ytest)
            epoch_count.append(epoch)
            loss_vals.append(loss.item())           # .item() ✅
            test_loss_vals.append(test_loss.item()) # .item() ✅
            print(f"epoch={epoch} | loss: {loss.item():.4f} | test_loss: {test_loss.item():.4f}")
```
