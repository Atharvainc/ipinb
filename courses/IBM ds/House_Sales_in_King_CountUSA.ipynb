```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
import sklearn as skl
```

```python
df=pd.read_csv('kc_house_data_NaN.csv')
df.head()
```

```python
#1
df.dtypes
```

```python
#2
df.drop('Unnamed: 0',inplace=True,axis=1)
df.drop('id',inplace=True,axis=1)
df.describe()
```

```python
mean=df['bedrooms'].mean()
df['bedrooms']=df['bedrooms'].replace(np.nan,mean)
mean=df['bathrooms'].mean()
df['bathrooms']=df['bathrooms'].replace(np.nan,mean)
```

```python
#3
floorcount=df['floors'].value_counts()
floorsdf=floorcount.to_frame()
floorsdf
```

```python
#4
sbn.boxplot(x='waterfront',y='price',data=df)
```

```python
#5
sbn.regplot(x='sqft_above',y='price',data=df)
plt.ylim(0,)
#+ve correlation
```

```python
#6
x=df[['sqft_living']]
y=df['price']
lr=skl.linear_model.LinearRegression()
lr.fit(x,y)
lr.score(x,y)
```

```python
#7
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]
w=df[features]
lr.fit(w,y)
lr.score(w,y)
```

```python
#8
Input=[('scale',skl.preprocessing.StandardScaler()),('polynomial',skl.preprocessing.PolynomialFeatures(include_bias=False)),('model',skl.linear_model.LinearRegression())]
pipe=skl.pipeline.Pipeline(Input)
pipe.fit(w,y)
pipe.score(w,y)
```

```python
#9
xtrain,xtest,ytrain,ytest=skl.model_selection.train_test_split(w,y,test_size=0.15,random_state=1)
rm=skl.linear_model.Ridge(alpha=0.1)
rm.fit(xtrain,ytrain)
rm.score(xtest,ytest)
```

```python
#10
pol=skl.preprocessing.PolynomialFeatures(degree=2,include_bias=False)
xtestpoly=pol.fit_transform(xtest)
xtrainpoly=pol.fit_transform(xtrain)
rm.fit(xtrainpoly,ytrain)
ycap=rm.predict(xtestpoly)
skl.metrics.r2_score(ytest,ycap)
```
