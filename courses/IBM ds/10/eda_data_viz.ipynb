```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
import requests
import io
url='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv'
dihta=requests.get(url)
df=pd.read_csv(io.StringIO(dihta.text))
df.head()
```

```python
sbn.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
sbn.catplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
sbn.catplot(y="PayloadMass", x="LaunchSite", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sbn.barplot(x='Orbit',y='Class',data=df,errorbar=None)
plt.xticks(rotation=90)
plt.show()
```

```python
sbn.catplot(y='Orbit', x="FlightNumber",hue='Class', data=df,aspect=1.78)
plt.show()
```

```python
sbn.catplot(y='Orbit', x="PayloadMass",hue='Class', data=df,aspect=1.78)
plt.show()
```

```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
```

```python
sbn.lineplot(x='Date',y='Class',data=df)
plt.show()
```

```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

```python
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])
features_one_hot.head()
```

```python
features_one_hot = features_one_hot.astype(float)
```
