```python
class Animal:
    species_type = "Unknown"

    def __init__(self, name, sound):
        self.name = name
        self.sound = sound
        print(f"Animal object {self.name} created.")

    def speak(self):
        print(f"{self.name} says {self.sound}")
```

## inheritance
* single
* multilevel
* heirarchal
* multiple
* hybrid

```python
class Dog(Animal):
    def __init__(self, name):
        super().__init__(name, "Woof")
    
    def speak(self):
        print(f"{self.name} barks: Woof!")
        
    def fetch(self):
        print(f"{self.name} is fetching.")

d1 = Dog("Buddy")
d1.speak()
d1.fetch()
```

```python
class Puppy(Dog):
    def __init__(self, name):
        super().__init__(name)
    
    def speak(self):
        print(f"{self.name} yips: yip yip!")

p1 = Puppy("Max")
p1.speak()
p1.fetch()
```

```python
class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, "Meow")
    
    def speak(self):
        print(f"{self.name} meows: Meow!")
        
    def climb(self):
        print(f"{self.name} is climbing.")

c1 = Cat("Whiskers")
c1.speak()
c1.climb()
print("(Dog and Cat both inherit from Animal)")
```

```python
class Walker:
    def walk(self):
        print("I am walking.")

class Swimmer:
    def swim(self):
        print("I am swimming.")

class Duck(Animal, Walker, Swimmer):
    def __init__(self, name):
        super().__init__(name, "Quack")
        
    def speak(self):
        print(f"{self.name} quacks: Quack!")

duck1 = Duck("Donald")
duck1.speak()
duck1.walk()
duck1.swim()

```

```python
class Pet:
    def play(self):
        print("Playing with owner.")
        
class PetDog(Dog, Pet):
    def __init__(self, name):
        super().__init__(name)
        
pd1 = PetDog("Rover")
pd1.speak()
pd1.fetch()
pd1.play()

```

## ploymorphism

```python

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
        
    def __str__(self):
        return f"Point({self.x}, {self.y})"

pt1 = Point(1, 2)
pt2 = Point(5, 3)
pt3 = pt1 + pt2

print(f"{pt1} + {pt2} = {pt3}")

v4 = Point(6, 5)
print(f"Is {pt3} equal to {v4}? {pt3.x == v4.x and pt3.y == v4.y}")
print(f"Is {pt1} equal to {v4}? {pt1.x == v4.x and pt1.y == v4.y}")
```
