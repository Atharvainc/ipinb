```python
class Student:
    school_name = "djsce"
    def __init__(self, name, student_id, grade):
        self.name = name
        self.student_id = student_id
        self.grade = grade
        
    def display_details(self):
        print(f"\n--- Student Details ---")
        print(f"Name: {self.name}")
        print(f"ID: {self.student_id}")
        print(f"Grade: {self.grade}")
        print(f"School: {self.school_name}")

    def update_grade(self, new_grade):
        self.grade = new_grade
        print(f"\n{self.name}'s grade has been updated to {self.grade}.")

student1 = Student("Adeep Roshan", "c001", 10)
student2 = Student("Paarth dubey", "c012", 9)
student1.display_details()
student2.display_details()
print(f"\nAccessing an attribute directly: {student1.name}")
student1.update_grade(11)
student1.display_details()
print(f"\nAccessing class variable via class: {Student.school_name}")
print(f"Accessing class variable via instance: {student1.school_name}")
Student.school_name = "spit"
print("\n School name changed! ")
student1.display_details()
student2.display_details()
```
